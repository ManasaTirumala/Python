```
import numpy as np, pandas as pd
df=pd.read_csv('Service_Area_PUF.csv',usecols=['StateCode','StateCode2','PartialCounty','IssuerId2','CoverEntireState'])
df


```

```

#Get top X  categories in the selected column i.e colName
def getTop_X_cat(X,colName):
  for colName in df.columns:
       if colName in df:
          freq=df[colName].value_counts().sort_values(ascending=False)
          top =[x for x in freq.head(X).index] 
          print(freq)
          return top
top_X=getTop_X_cat(10,'StateCode2')
# print(top_X)

# #Apply one hot encoding for the top X categorical variables in each column
def oneHotEncode_topX(df,colName,top_X):
  for i in top_X:
    df[colName+'_'+i]=np.where(df[colName]==i,1,0)

oneHotEncode_topX(df,"StateCode2",top_X)
  
df.head(10)
```
