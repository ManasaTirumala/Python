```
import pandas as pd
import numpy as np
df= pd.read_csv('BenzData.csv', usecols=['X1','X2','X3','X4','X5','X6','X8'])
df.head(100)

```

```
for cat in df.columns:
   print(cat, '::', len(df[cat].unique()),"labels")
pd.get_dummies(df,drop_first=True).shape
```

```
 freq=df.X1.value_counts().sort_values(ascending=False)
 top_10=[x for x in freq.head(10).index] 
 top_10
```

```
for label in top_10:
        df[label]= np.where(df.X1==label,1,0)
df[['X1']+top_10].head(10)
 
  
```

```
def one_hot_topX(df, top_x,col):
  for label in top_x:
    df[col+'_'+label]=np.where(df[col]==label,1,0)
df= pd.read_csv('BenzData.csv', usecols=['X1','X2','X3','X4','X5','X6','X8'])
one_hot_topX(df,top_10,'X3')
df.head(30)
```
